```python
weather = weather.drop(columns=['STATION', 'AWND', 'FMTM', 'PGTM', 'WDF2', 'WDF5', 'WSF2', 'WSF5', 'WT06', 'WT07', 'WT08', 'WT10', 'WT11', 'WT13', 'WT17', 'WT18', 'WT19'])
weather
```

```python
keys = weather.columns.tolist()

values = ['date', 'precipitation', 'snowfall', 'snow_depth', 'avg_temp', 'max_temp', 'min_temp', 'foggy', 'heavy_fog', 'thunder', 'hail', 'blowing_snow', 'rain']

col_name_dict = dict(zip(keys, values))

weather = weather.rename(columns=col_name_dict)
weather
```

```python
weather.avg_temp = weather.avg_temp.fillna((weather.max_temp + weather.min_temp) / 2)
```

```python
weather
```

```python
boolean_columns = ['foggy','heavy_fog','thunder','hail','blowing_snow','rain']

for column in boolean_columns:
    weather[f'{column}'] = weather[f'{column}'].fillna(0).astype(int)
    
weather
```

```python
weather.date = pd.to_datetime(weather.date)
```

```python
weather.info()
```

```python
weather = weather.sort_values("date").set_index("date")
weather
```

```python
weather.precipitation.rolling(7).mean()
```

```python
rolling_periods = [7, 14, 30]
features = ['precipitation', 'avg_temp', 'max_temp', 'min_temp']

for feature in features:
    for period in rolling_periods:
        weather[f'{feature}_rolling_{period}'] = weather[f'{feature}'].rolling(period).mean()
        
weather
```
